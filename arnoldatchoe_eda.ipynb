BRIEF DESCRIPTION OF DATASET
This project makes use of the "taxis" dataset included in the seaborn library. A copy of the csv_file containing this dataset
was obtained from https://github.com/mwaskom/seaborn-data/blob/master/taxis.csv. The csv_file has been copied to our project repository.
Dataset contains information on yellow and green taxi trips in New York city for march of 2019. The dataset was collected by various technology providers under the Taxicab and Livery Passenger Enhancement Programs(TPEP/LPEP) and made available to NYC Taxi an Limousine Commission. The dataset include features such as pickup and drop-off date/time, pickup and drop-off location, trip distance,itemized fares, rate types, payment types and passenger count reported by drivers. This brief description of the project dataset was obtained from https://www.nyc.gov/site/tlc/about/tlc-trip-record-data.page.

FEATURE, ATTRIBUTE OR COLUMN DESCRIPTION
1. pickup: This feature provides information on the date and time passengers were pick a taxi at their pickup location.
2. dropoff: Date and time passengers were dropped off at their various dropoff locations.
3. passenger Count: The number of passengers that used the taxi service at each instance, indicated by the driver.
4. distance: Distance in miles between pickup location and dropoff location.
5. fare: The base fare of the trip.
6. tip: The amount of tip given to the driver.
7. toll: The toll fees incurred for each trip.
8. totals: Total cost of trips which includes fare, tip and toll.
9. color: This feature represents the color of each taxi observed.
10. payment: This represents the type of payment used by the passengers to settle their fees.
11. pickup_zone: The exact location passengers were picked up.
12. dropoff_zone: The exact location passengers were dropped off.
13. pickup_borough:The borough each pickup_zone is located in.
14. dropoff_borough: The borough each dropoff_zone is locate in.
Note: Boroughs represent major governmental districts in New York City.

STEP 6: INITIAL DATA TRANSFORMATION AND FEATURE ENGINEERING
Drop pickup column.
Drop dropoff column.
Drop the old total column and create a new total_charge column which is the sum of fare, tolls, and tips.
Rename passenger to passenger_count.
Rename distance to distance_miles.
Rename payment to payment_type.

```python
# Drop columns identified in the notes above.
df2 = df.drop(columns=['pickup','dropoff','total'])

# Rename columns identified in the notes above.
df3 = df2.rename(columns={'passengers':'passenger_count', 'distance':'distance_miles', 'payment':'payment_type'})

# Verify changes have be implemented.
df3.head()
```

```python
# Create a new total_charge column. The sum of fare, tolls and tips.
df3['total_charge'] = df3['fare']+df3['tip']+df3['tolls']

# Verify changes has been implemented.
df3.head()
```

STEP 7: INITIAL VISUALIZATION
Visualize key aspects of the taxi dataset. Summarize and extract need information.

GOAL: Determine revenue for march,2019 for each of the two types of payment used and taxi color.
CHART TYPE: Bar chart.

```python
# Summarize revenue by payment type and color.
summary8 = df3.groupby(['payment_type','color'])['total_charge'].sum().reset_index(name='revenue')
print(summary8)

plt.figure(figsize=(10, 6))
sns.barplot( data = summary8, x='payment_type', y='revenue',hue= 'color')

# Add titles and labels
plt.title('Revenue Per Payment Type', fontsize=16)
plt.xlabel('Payment Type', fontsize=14)
plt.ylabel('Revenue', fontsize=14)
plt.legend(title = 'Taxi color')
```

OBSERVATION
1. Revenue Sources: The majority of revenue in March 2019 was generated through credit card payments, totaling approximately $77,154, compared to about $21,346 from cash payments. This indicates a strong preference for electronic payment methods among customers.

2. Taxi Type Revenue: Yellow taxis significantly outperformed green taxis in revenue generation. This discrepancy can be attributed to the operational capabilities of yellow taxis, which are allowed to pick up passengers from anywhere in New York City, while green taxis are restricted to specific areas.
